```python
import requests
from bs4 import BeautifulSoup
import re

def get_wikipedia_links(url):
    # Send GET request and get HTML content
    response = requests.get(url)
    
    # Create BeautifulSoup object
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all links on the page
    all_links = soup.find_all('a')
    
    # Filter for Wikipedia links
    wikipedia_links = []
    for link in all_links:
        href = link.get('href', '')
        
        # Check if it's a Wikipedia URL
        if href.startswith('/wiki/'):
            # Remove any query parameters
            cleaned_url = re.sub(r'\?.*$', '', href)
            wikipedia_links.append(cleaned_url)
    
    return set(wikipedia_links)

# Example usage
url = 'https://en.wikipedia.org/wiki/Li_Cong_(taikonaut)'
wikipedia_links = get_wikipedia_links(url)

print(f"Number of Wikipedia links found: {len(wikipedia_links)}")
for link in wikipedia_links:
    print(link)
```
