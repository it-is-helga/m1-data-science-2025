### Question 1.1

- Look up http://dbpedia.org/page/Alan_Bean. Write a query to retrieve the abstract.


````python
from SPARQLWrapper import SPARQLWrapper, JSON
sparql = SPARQLWrapper("http://dbpedia.org/sparql/")
sparql.setQuery("""
  PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
  SELECT ?label
  WHERE {
  <http://dbpedia.org/resource/Alan_Bean> rdfs:label ?label }
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
for result in results["results"]["bindings"]:
    print(result["label"]["value"], result["label"]["xml:lang"])
````

````python
sparql.setQuery(qcit)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
results 
````

````python
qastro = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
 PREFIX dbp: <http://dbpedia.org/ontology/>
 SELECT ?astronaut WHERE {
  ?astronaut dbo:occupation dbr:Astronaut .
}
"""

sparql.setQuery(qastro)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
results 
````

### Question 1.5
- Retrieve all astronauts and their nationality: Try with and without the OPTIONAL clause; for each query, get the number of astronauts.


````python
q15 ="""
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX dbp: <http://dbpedia.org/ontology/>
SELECT ?astronaut ?nationality WHERE {
  ?astronaut dbo:occupation dbr:Astronaut .
  ?astronaut dbo:nationality ?nationality .
}
"""

sparql.setQuery(q15)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
results 
````

````python
q151 = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX dbp: <http://dbpedia.org/ontology/>
SELECT (COUNT(?astronaut) AS ?numAstronauts) WHERE {
  ?astronaut dbo:occupation dbr:Astronaut .
  ?astronaut dbo:nationality ?nationality .
}
"""

sparql.setQuery(q151)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
results 
````

````python
q152 ="""
SELECT ?astronaut ?nationality WHERE {
  ?astronaut dbo:occupation dbr:Astronaut .
  OPTIONAL { ?astronaut dbo:nationality ?nationality . }
}
GROUP BY ?astronaut ?nationality
"""

sparql.setQuery(q152)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
results 
````

````python
q1522 = """
SELECT (COUNT(?astronaut) AS ?numberOfAstronauts) WHERE {
  ?astronaut dbo:occupation dbr:Astronaut .
}
"""

sparql.setQuery(q1522)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
results 
````
